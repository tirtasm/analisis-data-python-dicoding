### Cleaning Data

kita akan menghapus tabel yang digunakan karena kedua tabel itu terlihat sama hanya berbeda di kolom hr saja, 
jadi akan menghapus tabel days karena tidak ada hr sehingga tidak relevan dengan pertanyaan bisnis kita membahas waktu terbanyak dalam sehari


```python
del days_df
```

selanjutnya akan menghapus column yang tidak diperlukan
yang akan dihapus yaitu 
- windspend karena tidak ada hubungannya dengan kecepatan angin dengan penyewa sepeda
- instant karena tidak ada informasi yang berguna 

```python
hour_df.drop(columns=["instant", "windspeed"], inplace=True)

hour_df.head()
```

selanjutnya akan mengubah nama kolom dan tipe data agar mudah dibaca

```python
hour_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'hr': 'hour',
    'weathersit': 'weather',
    'cnt': 'count'
}, inplace=True)
hour_df['dateday'] = pd.to_datetime(hour_df.dateday)
hour_df.head()
```

mengubah angka menjadi string

```python
hour_df['month'] = hour_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6:'Jun', 7:'Jul', 8:'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
hour_df['weekday'] = hour_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
hour_df['season'] = hour_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
hour_df['weather'] = hour_df['weather'].map({
   1: 'Sunny/Cloudy', 2: 'Foggy/Deep Cloudy', 3: 'Light Rain/Snow', 4: 'Extreme Weather'
})
```

```python
hour_df['season'] = hour_df.season.astype('category')
hour_df['year'] = hour_df.year.astype('category')
hour_df['month'] = hour_df.month.astype('category')
hour_df['holiday'] = hour_df.holiday.astype('category')
hour_df['weekday'] = hour_df.weekday.astype('category')
hour_df['workingday'] = hour_df.workingday.astype('category')
hour_df['weather'] = hour_df.weather.astype('category')
hour_df.head() 
```

## Exploratory Data Analysis (EDA)

```python
#melihat lagi info dari data hour
hour_df.info()
```

```python
#melihat lagi statistik dari data hour
hour_df.describe(include='all')
```

terlihat tidak ada keanehan kembali

```python
# mengelompokkan penyewa sepeda dari kondisi cuaca
hour_df.groupby(by='weather', observed=False).agg({
     'count': ['max', 'min', 'mean', 'sum']    
})
```

dari hasil diatas, bahwa paling banyak pengguna sepeda cenderung lebih sering disewa saat cuaca cerah atau berawan dan paling sedikit pengguna sepeda menyewa saat cuara buruk

```python
#mengelompokkan penyewa sepeda d|ari jam 
hour_df.groupby(by='hour').agg({
    'count': ['max', 'min', 'mean', 'sum']
}).sort_values(by=('count', 'sum'), ascending=False)
```

dari hasil diatas, bahwa paling banyak pengguna sepeda lebih sering menyewa pada saat pukul 17.00 dan paling sedikit pengguna sepeda menyewa saat pukul 04.00

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah cuaca mempengaruhi jumlah penyewaan sepeda?

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='weather', y='count', data=hour_df, hue='weather')
plt.title('Pengaruh cuaca  pada jumlah penyewaan sepeda')
plt.xlabel('Cuaca')
plt.ylabel('Jumlah pnyewaan')
plt.show()
```

### Pertanyaan 2: Kapan waktu penyewaan sepeda paling banyak dalam sehari?

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='hour', y='count', data=hour_df, color='r')
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Jam dalam Sehari')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(range(0, 24))
plt.grid()
plt.show()
```

**Insight:**
- Berdasarkan tabel batang tersebut, penyewa cenderung lebih memilih saat cuaca cerah karena penyewa nyaman menggunakannya daripada cuaca yang buruk karena cuaca akan mengganggu kenyamanan penyewa
- Berdasarkan garis data, penyewaan sepeda paling banyak terjadi pada pukul 17.00 dan pukul 18.00, kemudian untuk penyewa sepeda paling sedikit pada pukul 04.00. Mungkin orang penyewa sepedanya banyak yang menyewa setelah jam kerja atau pulang kerja dan di saat pagi hari mungkin orang tidak menggunakan sepeda karena bisa menggunakan transportasi lain seperti bus, kereta, dan lain lain

## Conclusion

Kapan waktu penyewaan sepeda paling banyak dalam sehari?
- Penyewaan sepeda paling banyak terjadi pada pukul 17.00 dan 18.00. Karena kemungkinan besar waktu tersebut bertepatan dengan jam pulang kerja atau aktivitas sore hari, di mana mereka lebih cenderung menggunakan sepeda untuk pulang dari kantor. Sebaliknya, penyewaan sepeda paling sedikit terjadi pada pukul 04.00, karena waktu tersebut masih pagi hari, sehingga aktivitas pengguna cenderung sangat sedikit
  
Apakah cuaca memengaruhi jumlah penyewaan sepeda?
- Berdasarkan visualisasi, penyewaan sepeda cenderung lebih banyak pada cuaca cerah dibanding cuaca buruk. Hal ini masuk akal karena kondisi cuaca buruk seperti hujan dapat mengurangi kenyamanan pengguna sepeda dan bahkan meningkatkan risiko kecelakaan.
Dengan demikian, cuaca yang cerah menjadi faktor yang pilihan dari tingginya jumlah penyewaan sepeda.
